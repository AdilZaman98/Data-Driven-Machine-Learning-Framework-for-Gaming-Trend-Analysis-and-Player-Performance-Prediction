# **Outliers of the Dataset:**

```python
# Sirf numeric columns select karein
numeric_cols = df.select_dtypes(include=['number']).columns

# Loop through each numeric column
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Outliers count karein
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    
    if outlier_count > 0:
        print(f"Column '{col}' has {outlier_count} outliers.")
```

# **Predictive Pricing Intelligence: Comparative Regression Analysis for Gaming Markets:**

>**The refined objective of this script is to build a high-precision Price Prediction Engine by evaluating various Machine Learning models against gaming-specific market data. It utilizes ML Regressors to identify the specific algorithm—such as Lasso or Linear Regression—that most accurately captures the relationship between hardware specs and market value. This comparative approach ensures the final model minimizes pricing errors (RMSE) while maximizing predictive reliability ($R^2$) for real-world valuation..**

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import r2_score, mean_squared_error,root_mean_squared_error
from sklearn.pipeline import Pipeline
def load_data(path="Gaming.csv", target_col=None):
    """
    Load CSV and return X, y.
    By default attempts to infer target column:
      - if target_col provided, uses that
      - else if a column named 'price' exists, uses 'price'
      - else prompts user to set target_col (here raise)
    """
    df = pd.read_csv(path)
    if target_col is None:
        if "price" in df.columns:
            target_col = "price"
        else:
            raise ValueError("Target column not specified and 'price' not found. Provide target_col.")
    # drop non-numeric columns except those convertible; keep numeric features only for simplicity
    X = df.drop(columns=[target_col])
    # Basic preprocessing: drop columns with too many missing or non-numeric; keep numeric dtype columns
    numeric = X.select_dtypes(include=[np.number]).copy()
    # If no numeric columns found, attempt to extract simple numeric from strings (not done here)
    if numeric.shape[1] == 0:
        raise ValueError("No numeric features detected. Please preprocess dataset to produce numeric features (e.g., extract specs).")
    # drop rows with missing target or numeric NaNs
    data = pd.concat([numeric, df[target_col]], axis=1).dropna()
    y = data[target_col].values
    X = data.drop(columns=[target_col]).values
    return X, y, numeric.columns.tolist()

def compare_models(X, y, feature_names=None, random_state=42):
    """
    Trains multiple regressors and returns a DataFrame with metrics.
    Uses a train/test split plus cross-val (5-fold) R2 for robustness.
    """
    models = {
        "LinearRegression": LinearRegression(),
        "Ridge": Ridge(random_state=random_state),
        "Lasso": Lasso(random_state=random_state),
        "DecisionTree": DecisionTreeRegressor(random_state=random_state),
        "RandomForest": RandomForestRegressor(n_estimators=200, random_state=random_state),
        "GradientBoosting": GradientBoostingRegressor(n_estimators=200, random_state=random_state),
        "SVR": SVR(kernel="rbf"),
        "KNeighbors": KNeighborsRegressor(n_neighbors=5),
        "MLP": MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, random_state=random_state)
    }

    # train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

    results = []
    # We'll standardize features for models that need it (SVR, MLP, KNN, Ridge, Lasso)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    # For tree-based models we use unscaled features
    for name, model in models.items():
        use_scaled = name in {"SVR", "MLP", "KNeighbors", "Ridge", "Lasso", "LinearRegression"}
        Xtr = X_train_scaled if use_scaled else X_train
        Xte = X_test_scaled if use_scaled else X_test

        # Fit
        model.fit(Xtr, y_train)
        # Predict
        preds = model.predict(Xte)
        # Metrics
        r2 = r2_score(y_test, preds)
        mse = mean_squared_error(y_test, preds)
        rmse = np.sqrt(mse)
        # cross-validated R2 on full data (5-fold)
        kf = KFold(n_splits=5, shuffle=True, random_state=random_state)
        # Use same scaling logic for cross-val: wrap scaler+model in pipeline if needed
        if use_scaled:
            pipe = Pipeline([("scaler", StandardScaler()), ("model", model.__class__(**(model.get_params() if hasattr(model,'get_params') else {})))])
            try:
                cv_scores = cross_val_score(pipe, X, y, scoring="r2", cv=kf, n_jobs=-1)
                cv_mean = np.mean(cv_scores)
            except Exception:
                cv_mean = np.nan
        else:
            try:
                cv_scores = cross_val_score(model, X, y, scoring="r2", cv=kf, n_jobs=-1)
                cv_mean = np.mean(cv_scores)
            except Exception:
                cv_mean = np.nan

        results.append({
            "model": name,
            "r2_test": r2,
            "rmse_test": rmse,
            "cv_r2_mean": cv_mean
        })

    res_df = pd.DataFrame(results).sort_values(by="r2_test", ascending=False).reset_index(drop=True)
    return res_df

def plot_results(res_df, metric="r2_test"):
    """
    Bar chart comparing models by chosen metric ('r2_test' or 'rmse_test' or 'cv_r2_mean').
    """
    plt.figure(figsize=(10,6))
    sns.set_style("whitegrid")
    order = res_df.sort_values(by=metric, ascending=(metric!="r2_test"))["model"]
    if metric == "rmse_test":
        sns.barplot(x="model", y=metric, data=res_df, order=order, palette="viridis")
        plt.ylabel("RMSE (lower = better)")
    else:
        sns.barplot(x="model", y=metric, data=res_df, order=order, palette="viridis")
        plt.ylabel(metric.replace("_", " ").upper() + " (higher = better)")
    plt.xlabel("Model")
    plt.xticks(rotation=45, ha="right")
    plt.title(f"Model comparison by {metric}")
    plt.tight_layout()
    plt.show()

def main():
    # Path to your CSV file; change if needed
    csv_path = "Gaming.csv"
    # target_col: change if you want a different target
    target_col = "price"

    print("Loading data...")
    X, y, feats = load_data(csv_path, target_col=target_col)
    print(f"Numeric features used ({len(feats)}): {feats}")
    print(f"Samples: {X.shape[0]}, Features: {X.shape[1]}")

    print("Comparing models...")
    res_df = compare_models(X, y, feature_names=feats)
    print(res_df)

    # Plot R2 (test) and RMSE
    plot_results(res_df, metric="r2_test")
    plot_results(res_df, metric="rmse_test")

    # Report best model by r2_test
    best = res_df.iloc[0]
    print(f"\nBest model by test R²: {best['model']} (R²={best['r2_test']:.4f}, RMSE={best['rmse_test']:.4f}, CV_R2={best['cv_r2_mean']:.4f})")

if __name__ == "__main__":
    main()
```
